```python
imgh=img.shape[0]
imgw=img.shape[1]
px=0
py=0
```

```python
def click_event(event, x, y, flags, params):
 
    # checking for left mouse clicks
    if event == cv2.EVENT_LBUTTONDOWN:
 
        # displaying the coordinates
        # on the Shell
        print(x, ' ', y)
        px=x
        py=y
 
        # displaying the coordinates
        # on the image window
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(img, str(x) + ',' +
                    str(y), (x,y), font,
                    1, (255, 0, 0), 2)
        cv2.imshow('image', img)
 
    # checking for right mouse clicks    
    if event==cv2.EVENT_RBUTTONDOWN:
 
        # displaying the coordinates
        # on the Shell
        print(x, ' ', y)
        px=x
        py=y
 
        # displaying the coordinates
        # on the image window
        font = cv2.FONT_HERSHEY_SIMPLEX
        b = img[y, x, 0]
        g = img[y, x, 1]
        r = img[y, x, 2]
        cv2.putText(img, str(b) + ',' +
                    str(g) + ',' + str(r),
                    (x,y), font, 1,
                    (255, 255, 0), 2)
        cv2.imshow('image', img)
 

 
img = cv2.imread("F:/kuet/4th year/4.1/CSE 41/CSE 4128/5.2/fill.jpg", 0)
cv2.imshow('image', img)   
cv2.setMouseCallback('image', click_event)  
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
px=92
py=251

array=np.zeros((imgh,imgw),np.uint8)
array[py,px]=255
plt.imshow(array , cmap="gray")
```

```python
kernel= cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
kernel
    
```

```python
imgcomp=255-img
plt.imshow(imgcomp , cmap="gray")
```

```python
while True:
    temp=array
    array=cv2.dilate(array,kernel)
    array=cv2.bitwise_and(array,imgcomp)
    if cv2.countNonZero(cv2.subtract(array,temp))==0:
        break
```

```python
plt.imshow(array , cmap="gray")
```

```python
img2=cv2.bitwise_or(img,array)
plt.imshow(img2 , cmap="gray")
```
